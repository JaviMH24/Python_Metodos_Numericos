# **Funciones**

En programación, una ``función`` es una secuencia de instrucciones que realiza una tarea específica. Una función es un bloque de código que se ejecuta al ser llamada. Una función puede tener argumentos de entrada, que son proporcionados por el usuario, la entidad que la llama. Las funciones también tienen parámetros de salida, que son los resultados que el usuario espera recibir una vez que la función haya completado su tarea. 

```python
# Definir una funcion
def saludar(nombre):
    print(f"Hola, {nombre}!")

saludar("Mundo")
```

```python
def suma(a,b):
    return a + b

resultado = suma(3, 5)
print(resultado)
```

Cuando se realizan funciones complejas y deseamos saber que es lo que hace, a menudo dentro de la funcion se escribe un bloque de comentarios que ayuda a saber de que ayuda la función. Para llamar que es lo que hace la función utilizamos la función ``help ()``

```python
def multiplicar(a,b):
    """Multiplica dos números y devuelve el resultado."""
    return a * b

help(multiplicar)
```

# **Variables globales y locales**

Es imporante reconocer que variables s eencuentran definidas como globales y locales, ya que pueden llevar a resultados distintos, como este ejemplo:


```python
#En esta funcion la variable out es una variable local
def my_adder(a, b, c):
    out = a + b + c
    print(f'The value out within the function is {out}')
    return out

#Aqui el valor de out es una variable global
out = 1
d = my_adder(1, 2, 3)
print(f'The value out outside the function is {out}')
```

Es importante que si utilizas un valor global y despues lo mandas a llamar dentro de la función, debes de llamarlo con la función ``global`` para que python la reconozca, observa el siguiente ejemplo:

```python
n = 42

def func():
    print(f'Within function: n is {n}')
    n = 3
    print(f'Within function: change n to {n}')

func()
print(f'Outside function: Value of n is {n}')
```

```python
n = 42

def func(): 
    global n #Esta linea le dice a Python que n es una variable global, no local para que puede ser modificada dentro de la funcion
    print(f'Within function: n is {n}')
    n = 3
    print(f'Within function: change n to {n}')

func()
print(f'Outside function: Value of n is {n}')
```

# **Funciones anidadas**

Una ``función anidada`` es aquella que se define dentro de otra función (la función padre). Solo la función padre puede llamar a la función anidada. Sin embargo, la función anidada conserva un bloque de memoria independiente de su función padre.

Considere la siguiente función anidada:

```python
import numpy as np

def my_dist_xyz(x, y, z):
    """
    x, y, z are 2D coordinates contained in a tuple
    output:
    d - list, where
        d[0] is the distance between x and y
        d[1] is the distance between x and z
        d[2] is the distance between y and z
    """
    
    def my_dist(x, y):
        """
        subfunction for my_dist_xyz
        Output is the distance between x and y, 
        computed using the distance formula
        """
        out = np.sqrt((x[0]-y[0])**2+(x[1]-y[1])**2)
        return out
    
    d0 = my_dist(x, y)
    d1 = my_dist(x, z)
    d2 = my_dist(y, z)
    
    return [d0, d1, d2]
```

```python
d = my_dist_xyz((0, 0), (0, 1), (1, 1)) #se colocan los argumentos en tipo tupla ya que la funcion
                                        #espera tuplas como argumentos
print(d)
d = my_dist((0, 0), (0, 1))
```

Como podemos observar, si llamamos a la subfunción ``my_dist()`` mandar un error, en la que se establece que no esta definida. Esto sucede porque solo se puede acceder a la función padre y no a la subfunción.

# **La función lambda**

A veces, no queremos usar la forma habitual de definir una función, especialmente si nuestra función es de una sola línea. En este caso, podemos usar una función anónima en Python, que es una función que se define sin nombre.

Este tipo de funciones también se denominan funciones lambda, ya que se definen usando la palabra clave `labmda`. Una función lambda típica se define de la siguiente manera:


```python
#Nombramos a la función lambda como deseamos
raiz_cuadrada= lambda x: x**0.5 #Primero colocas el nombre de la función,
                                #luego el símbolo de igualdad, después la palabra clave lambda, 
                                # seguida de los argumentos separados por comas, dos puntos y 
                                # finalmente la expresión que se evalúa y se devuelve.

print(raiz_cuadrada(16))

#Utilizando la libreria numpy
import numpy as np

raiz_cuadrada = lambda x: np.sqrt(x)

print(raiz_cuadrada(16))
```

Escribe una función llamada `my_fun_plus_one` que reciba como argumentos de entrada un objeto función, `f`, y un número flotante `x`. `my_fun_plus_one` debe devolver `f` evaluada en `x`, y el resultado debe sumarse al valor 1. Verifica que funcione para diversas funciones y valores de `x`.

```python
import numpy as np 

def my_fun_plus_one(f, x):
    return f(x) + 1

print(my_fun_plus_one(np.sin, np.pi/2))
print(my_fun_plus_one(np.cos, np.pi/2))
print(my_fun_plus_one(np.sqrt, 25))
```
