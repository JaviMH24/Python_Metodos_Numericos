### - Cadenas de texto

Una cadena es una secuencia de caracteres, como "Hola Mundo". Las cadenas están rodeadas por comillas simples o dobles. Podríamos usar la función de print para enviar las cadenas a la pantalla.

```python
print("Hola mundo")
```

### Función type() y la función len()

Tambien se pueden asignar cadenas a variables, por ejemplo:

```python
g="Hello world"  
l='How are you?'

#Podemos saber que tipo de dato es con la función type()
print(type(g))
print(type(l))

#Podemos saber además la longitud de la cadena con la funcion len()
#La cadena comnienza desde la posicion 0. NOTA: Los ESPACIOS cuentan como caracter
print(len(g))
print(len(l))
```

### Acceder a un caracter especifico de la cadena o a un conjunto de caracteres

Podemos acceder a cualquier posicion de la cadena, colocando la variable donde se guardo la cadena y el valor de la posicion que deseamos obserar, por ejemplo:

```python
print(g[4])
print(l[8])
```

También podríamos seleccionar una secuencia usando el segmento de cadenas. Por ejemplo, si queremos obtener el "World", podríamos hacer el siguiente comando:

```python
print(g[6:11])
```

### Unir dos o más cadenas

```python
str_a = "I love Python! "
str_b = "You too!"

print(str_a + str_b)
```

# **Introduccioón a los arrays de Numpy**

Para crear arreglos a traves de numpy se utilizan los siguiente comandos:. Estos son ideales si deseamos representar matrices:

```python
import numpy as np

#areglo unidimensional 

x=np.array([1,2,3])
print(x)
```

Si deseamos agregar más arreglosm, debemos agregar otros corchetes 

```python
y=np.array([[4,5,6],[1,2,3]])
y
```

Muchas veces necesitamos conocer el tamaño o la longitud de una matriz.

- El atributo `shape` se aplica a una matriz M y devuelve una matriz de 2 × 3, donde el primer elemento es el número de filas de la matriz M.

- El segundo, el número de columnas. Cabe destacar que el resultado del atributo `shape` es una tupla. El atributo `size` se aplica a una matriz M y devuelve el número total de elementos de la matriz M.


```python
# Comando para conocer el numero de filas y columnas de un arreglo
y.shape
```

```python
#Devuelve el numero de elementos que contiene la matriz
y.size
```

Con frecuencia, necesitamos generar matrices con una estructura o patrón definido. Por ejemplo, podríamos querer crear la matriz z = [1 2 3 … 2000]. Sería muy engorroso escribir la descripción completa de z en Python. Para generar matrices ordenadas y con espaciado uniforme, resulta útil usar la función `arange` de NumPy.

```python
m=np.arange(1,10)  #Asi generamos un arreglo con los numeros del 1 al 9
                    #Podemos agregar el incremento con un tercer argumento, 
                    # por ejemplo np.arange(1,10,2) para obtener los numeros impares
m
```

```python
m=np.arange(1,10, 2) 
m
```

A veces queremos garantizar un punto de inicio y un punto final para un array, pero manteniendo elementos espaciados uniformemente.Por ejemplo, podríamos querer un array que empiece en 1, termine en 8 y tenga exactamente 10 elementos.

- Podemos usar la función ``np.linspace``. linspace toma tres valores de entrada separados por comas. Así,

A = linspace(a,b,n)

genera un array de n elementos espaciados uniformemente, que empieza en a y termina en b.


```python
A=np.linspace(0, 10, 5) #Genera un arreglo con 5 numeros equidistantes entre 0 y 10
A
```

Acceder a un elemento es parecido a las listas

```python
A[1]
```

```python
A[-1]
```

```python
A[2:]
```

# **Ingresar a elementos de matrices de 2 columnas**

Para acceder a los elementos de una matriz de 2 columnas, es necesario llamar a la coordenada en donde se encuentra, siguiendo el siguiente comando:

```python
x=np.array([[1,5,9],[3,5,7]])
x
```

```python
x[0,2] #Rcuerda que las filas y columnas comienzan desde la posición 0, 
        #por lo que x[0,2] se refiere al elemento en la primera fila y tercera columna del arreglo x.
```

Podemos acceder a todo el arreglo de la columna, con el siguiente comando:

```python
x[0, :]
```

```python
x[1:]
```

Podemos acceder a algunas filas y columnas especificas con el siguiente comando:

```python
x[[0,1],2] #Selecciona el elemento en la tercera columna de las filas 0 y 1 del arreglo x.
```

```python
x[:, 2] #Selecciona el elemento en la tercera columna de todas las filas del arreglo x.
```

# **Aritmetica de matrices**

Podemos generar operaciones entre matrices de la siguiete forma:

```python
B=np.array([[1, 2, 3], [4, 5, 6]])
B
```

```python
#Sumamos el valor 3 a cada elemento del arreglo B
B+3
```

```python
#Restamos el valor -6 a cada elemento del arreglo B  
B-6
```

```python
#Multiplicamos la matriz por 2
2*B
```

```python
#Dividimos entre 7  
B/7
```

```python
#Elevamos al cubo cada elemento del arreglo B
B**3
```

De la misma forma que en algebra lineal, podemos operar matrices, solo debemos tener cuidado en que las filas y columnas sigan las reglas de las matrices para operarlas.

```python
D=np.array([[1, 2], [-3, 14]])
D
```

Con este comando podemos aplicar la transupesta de la matriz:

```python
D.T
```

Algo increible de las matrics, es que podemos agregar condiciones logicas para operar con sus elementos, por ejemplo:

```python
x = np.array([1, 2, 4, 5, 9, 3])
y = np.array([0, 2, 3, 1, 2, 3])
```

```python
x>3 #se obtiene un arreglo de valores booleanos, donde cada elemento 
    #indica si la condición es verdadera o falsa para el elemento correspondiente en el arreglo x.
```

```python
x>y #se obtiene un arreglo de valores booleanos, donde cada elemento indica 
        #si la condición es verdadera o falsa para la comparación entre los elementos 
        # correspondientes en los arreglos x e y.
```

##### Modficar el array x, para que sus  sus elementos sean estrictamente mayores a 3

```python
x_modif=x[x>3] #se obtiene un nuevo arreglo que contiene solo los elementos de x que cumplen la condición x>3.
x_modif
```

Podmeos agregar una condicional para sustituir elementos por un numero dterminado:

```python
x[x>3]=2
x
```
