## Estructuras de datos

### Cadenas de texto

Una cadena es una secuencia de caracteres, como "Hola Mundo". Las cadenas están rodeadas por comillas simples o dobles. Podríamos usar la función de print para enviar las cadenas a la pantalla.

```python
print("Hola mundo")
```

### Función type() y la función len()

Tambien se pueden asignar cadenas a variables, por ejemplo:

```python
g="Hello world"  
l='How are you?'

#Podemos saber que tipo de dato es con la función type()
print(type(g))
print(type(l))

#Podemos saber además la longitud de la cadena con la funcion len()
#La cadena comnienza desde la posicion 0. NOTA: Los ESPACIOS cuentan como caracter
print(len(g))
print(len(l))
```

### Acceder a un caracter especifico de la cadena o a un conjunto de caracteres

Podemos acceder a cualquier posicion de la cadena, colocando la variable donde se guardo la cadena y el valor de la posicion que deseamos obserar, por ejemplo:

```python
print(g[4])
print(l[8])
```

También podríamos seleccionar una secuencia usando el segmento de cadenas. Por ejemplo, si queremos obtener el "World", podríamos hacer el siguiente comando:

```python
print(g[6:11])
```
