# **Recursion**

Una ``función recursiva`` es una función que se llama a sí misma. Funciona de forma similar a los bucles que describimos anteriormente, pero a veces es mejor usar recursión que bucles. Toda función recursiva tiene dos componentes: un ``caso base`` y un ``paso recursivo``.


- El ``caso base`` suele ser la entrada más pequeña y tiene una solución fácilmente verificable. Este es también el mecanismo que impide que la función se llame a sí misma indefinidamente.
 
- El ``paso recursivo`` es el conjunto de todos los casos en los que se realiza una llamada recursiva, es decir, una llamada de la función a sí misma.

### **La formula del factorial**

Por ejemplo, la función factorial se define como:

$$
f(n) =
\begin{cases}
1 & \text{si } n = 1  \\
n \times f(n - 1) & \text{en otro caso}
\end{cases}
$$

Notemos que si deseamos tranformar esta funcion en tipo recursion, el caso base seria el n=1 con resultado 1. Mientras que el paso recursivo seria en otro caso, y se manda a llamar a la funcion.

Esta funcion quedaria asi:


```python
def factorial(n):
    if n == 0 or n == 1: #caso base
        return 1
    else:
        return n * factorial(n - 1) #paso recursivo
    
factorial(6)
```

### **La sucesion de fibonacci**

Los números de Fibonacci se desarrollaron originalmente para modelar el crecimiento poblacional idealizado de los conejos. Desde entonces, se ha comprobado su relevancia en diversos fenómenos naturales.

Estos números se pueden generar mediante la siguiente fórmula recursiva. Cabe destacar que el paso recursivo contiene dos llamadas recursivas y que existen dos casos base (es decir, dos casos que detienen la recursión).

$$
f(n) =
\begin{cases}
1 & \text{si } n = 1  \\
1 & \text{si } n = 2  \\
F(n-1) + F(n - 2) & \text{en otro caso}
\end{cases}
$$

Y asi quedaria:

```python
def fibonacci(n):
    if n == 1: #caso base
        return 1
    elif n == 2: #caso base
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2) #paso recursivo
    
fibonacci(10)
```

Escribe una función llamada fibonacci_display que se base en la modificación de Fibonacci. ¿Puedes determinar el orden en que aparecerán los números de Fibonacci en la pantalla para fibonacci(5)?

```python
def fibonacci_display(n):
    if n == 1: # primer caso base
        out = 1
        print(out)
        return out
    elif n == 2: # segundo caso base
        out = 1
        print(out)
        return out
    else: 
        out = fibonacci_display(n-1)+fibonacci_display(n-2) # Paso recursivo
        print(out)
        return out # Recursive call 
    
fibonacci_display(5)
```

# **Funciones recursivas vs Iterativas**

En general, las funciones iterativas son más rápidas que las recursivas que realizan la misma tarea. 

La principal ventaja de escribir funciones recursivas es que generalmente se pueden escribir de forma mucho más compacta que las iterativas. El costo de esta mayor compacidad es un mayor tiempo de ejecución. 

Por ejemplo, si definimos a fibonnaci de estas dos formas, podremos ver el tiempo de ejecucion que toma cada una de ellas con el comando ``%timeit``

```python
def fibonacci(n):
    if n == 1: #caso base
        return 1
    elif n == 2: #caso base
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2) #paso recursivo
```

```python
import numpy as np

def iter_fib(n):
    fib = np.ones(n)
    
    for i in range(2, n):
        fib[i] = fib[i - 1] + fib[i - 2]
        
    return fib
```

```python
%timeit iter_fib(25)

%timeit fibonacci(25)
```

# **Divide y vencerás**

La estrategia de ``dividir y conquistar`` es muy útil para resolver problemas difíciles. Mediante esta estrategia, los problemas complejos se resuelven a partir de las soluciones de muchos problemas similares y sencillos. De esta forma, los problemas difíciles se dividen en partes más manejables.

En esta sección, analizaremos dos ejemplos clásicos de la estrategia de dividir y conquistar: el problema de las Torres de Hanoi y el algoritmo Quicksort.


### **Las Torres de Hanoi**

El problema de las ``Torres de Hanoi`` consiste en tres barras verticales, o torres, y N discos de diferentes tamaños, cada uno con un orificio en el centro para que la barra pueda deslizarse a través de él. Los discos están apilados inicialmente en una de las torres en orden descendente de tamaño (es decir, el disco más grande está en la base). El objetivo del problema es mover todos los discos a una barra diferente, cumpliendo las siguientes tres reglas:

- Solo se puede mover un disco a la vez. 
- Solo se puede mover el disco que se encuentra en la parte superior de la pila.
- No se puede colocar un disco encima de otro más pequeño. 

La clave del problema de las Torres de Hanoi reside en dividirlo en problemas más pequeños y fáciles de manejar, a los que llamaremos ``subproblemas``. En este caso, es relativamente fácil ver que mover un disco es sencillo (solo tiene tres reglas), pero mover una torre es difícil (no vemos inmediatamente cómo hacerlo). Por lo tanto, dividiremos el problema de mover una pila de tamaño ``N``en varios subproblemas que implican mover una pila de tamaño ``N − 1``.


### Solucion de forma recursiva

Consideremos una pila de N discos que deseamos mover de la Torre 1 a la Torre 3, y hagamos que `my_tower(N)` mueva una pila de tamaño N a la torre deseada (es decir, muestre los movimientos). La forma de escribir `my_tower` puede no ser evidente de inmediato. Sin embargo, si pensamos en el problema en términos de subproblemas, podemos ver que necesitamos mover los N-1 discos superiores a la torre central, luego el disco inferior a la torre derecha, y luego los N-1 discos de la torre central a la torre derecha. 

`my_tower` puede mostrar la instrucción para mover el disco N, y luego realizar llamadas recursivas a `my_tower(N-1)` para manejar el movimiento de las torres más pequeñas. Las llamadas a `my_tower(N-1)` realizan llamadas recursivas a `my_tower(N-2)`, y así sucesivamente. Un desglose de los tres pasos se muestra en la siguiente figura.

Y asi de simple queda la funcion:

```python
def my_towers(N, from_tower, to_tower, alt_tower):
    """
    Displays the moves required to move a tower of size N from the
    'from_tower' to the 'to_tower'. 
    
    'from_tower', 'to_tower' and 'alt_tower' are uniquely either 
    1, 2, or 3 referring to tower 1, tower 2, and tower 3. 
    """
    
    if N != 0:
        # recursive call that moves N-1 stack from starting tower
        # to alternate tower
        my_towers(N-1, from_tower, alt_tower, to_tower)
        
        # display to screen movement of bottom disk from starting
        # tower to final tower
        print("Move disk %d from tower %d to tower %d."\
                  %(N, from_tower, to_tower))
        
        # recursive call that moves N-1 stack from alternate tower
        # to final tower
        my_towers(N-1, alt_tower, to_tower, from_tower)
```

```python
my_towers(4, 1, 3, 2)
```

### **Quicksort u Ordenación rápida**

Una lista de números, A, se considera ordenada si sus elementos se encuentran en orden ascendente o descendente. Si bien existen muchos métodos para ordenar listas, el algoritmo Quicksort es una estrategia de divide y vencerás que resulta muy rápida para ordenar con un solo procesador (existen algoritmos más rápidos para múltiples procesadores).

El algoritmo Quicksort parte de la observación de que ordenar una lista es difícil, pero comparar es fácil. Por lo tanto, en lugar de ordenar la lista, la dividimos comparándola con un pivote. En cada llamada recursiva a Quicksort, la lista de entrada se divide en tres partes: elementos menores que el pivote, elementos iguales al pivote y elementos mayores que el pivote.

Luego, se realiza una llamada recursiva a Quicksort sobre los dos subproblemas: la lista de elementos menores que el pivote y la lista de elementos mayores que el pivote. Finalmente, los subproblemas se reducen lo suficiente (es decir, el tamaño de la lista es 1 o 0) como para que ordenar la lista sea trivial.


```python
def my_quicksort(lst):
    
    if len(lst) <= 1:
        # list of length 1 is easiest to sort 
        # because it is already sorted
        
        sorted_list = lst    
    else:
        
        # select pivot as teh first element of the list
        pivot = lst[0]
        
        # initialize lists for bigger and smaller elements 
        # as well those equal to the pivot
        bigger = []
        smaller = []
        same = []
        
        # loop through list and put elements into appropriate array
        
        for item in lst:
            if item > pivot:
                bigger.append(item)
            elif item < pivot:
                smaller.append(item)
            else:
                same.append(item)
        
        sorted_list = my_quicksort(smaller) + same + my_quicksort(bigger)
        
    return sorted_list
```

```python
my_quicksort([2, 1, 3, 5, 6, 3, 8, 10])
```
