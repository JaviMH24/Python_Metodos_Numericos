# **Formas de generar iteraciones**

## **- For Loops**

Un ``bucle for`` es un conjunto de instrucciones que se repite, o itera, para cada valor de una secuencia. A veces se les denomina ``bucles definidos`` porque tienen un inicio y un final predefinidos, delimitados por la secuencia.

El siguiente código muestra los primeros numeros naturales del 1 al 3

```python
n=0
for n in range(1,3):  #No toma el valor 3 solo el 1 y el 2
    print(n)
```

Tambien podemos crear una suma de estos numeros:

```python
n=0
for i in range (1,5):
    n=n+i
    print(n)
    
```

Podemos imprimi de forma simultanea cada caracter de una cadena de texto, de la siguiente manera:

```python
for c in "Python":
    print(c)
```

Tambien podemos imprimirlas desde una cadena ya arguemntada:


```python
s="Python"
for i in range(len(s)):  #len(s) devuelve el numero de caracteres de la cadena s
    print(s[i]) #s[i] devuelve el caracter de la cadena s en la posicion i
```

Dada una lista de enteros "a", sumar todos estos y guardarlos en una nueva variable.

```python
a=[1,2,3,4,5]
s=0

for i in range(len(a)):
    s=s+a[i]  #Tambien se puede escribir s+=a[i]
print(s)
```

Define un diccionario y recorre todas las claves y valores. Para generar esto, utilizaremos la palbra reservdada ``keys``, que viene definida para diccionarios.

```python
diccionario={"a":1,"b":2,"c":3}
s=0

for key in diccionario.keys():  #diccionario.keys() devuelve una lista con las claves del diccionario
    print(key, diccionario[key])  #diccionario[key] devuelve el valor asociado a la clave key
```

Además, otra alternativa es poder llamar a la palabra reservada `items`

```python
diccion1={"a":1, "b":3, "c":3}

for key, value in diccion1.items():
    print(key, value)  #diccion1.items() devuelve una lista de tuplas con las claves y los valores del diccionario

```

Otra forma es utilizar dos contadores en un bucle for, de la siguiente manera. Para ello se utilizará la función `zip`

```python
a={"a","b","c"}
b={12,15,20}

for i,j in zip(a,b):  #zip(a,b) devuelve una lista de tuplas con los elementos de a y b emparejados
    print(i,j)

```

### **Generar una función que verifica si existen digitos en una cadena**

```python
def mi_funcion(s):
    contador=0

    for c in s:
        if c.isdigit():  #isdigit() devuelve True si todos los caracteres de la cadena son dígitos
            contador=1
            break
        

    return contador
```

```python
mi_funcion("ghio3")
```

### **Generar una función que calcula la distancia ecuclidiana de dos puntos de la forma (x,y), utilizando bucles**

```python
import math

def my_dist_2_points(xy_points, xy):

    d = []  #Lista vacía para almacenar las distancias 
    for xy_point in xy_points:
        dist = math.sqrt(\
            (xy_point[0] - xy[0])**2 + (xy_point[1] - xy[1])**2)
        
        d.append(dist)
        
    return d

xy_points = [(1,2), (3,4), (5,6)]
xy = (2,3)
my_dist_2_points(xy_points, xy)
```
