# **Branching Statements**

Dentro de las funciones que podemos construir, existe la posibilidad de generar condicionales para que se evaluen ciertos pasos o condiciones necesarias para obtener el resultado deseado.

La condicional más utilizada es la condición ``if-else``


Probemos con la siguiente función que evalua la temperatura de un elemento. Si la condición de temperatura tiene una valor determinado se imprimirá la leyenda correspondiente.

```python
def temperature(mi_temperatura):
    if mi_temperatura < 0:
        return "Hace mucho frío"
    elif mi_temperatura < 15 and mi_temperatura >= 0: #podemos insertar condiciones lógicas para cada rango de temperatura
        return "Hace frío"
    elif mi_temperatura < 25 and mi_temperatura >= 15:
        return "Hace una temperatura agradable"
    elif mi_temperatura < 35 and mi_temperatura >= 25:
        return "Hace calor"
    else:
        return "Hace mucho calor"

temperature(19.9)
```

Función que identifica una clave correcta

```python
def clave(x):
    if x == "python":
        return "¡Clave correcta!"
    else:
        return "Clave incorrecta, inténtalo de nuevo."
    

password = input("Ingresa la contraseña: ")

clave(password)
```

Tambien podemos agregar condicionales que evaluen cadenas de texto:

```python
def is_odd(number):
    """
    function returns 'odd' if the input is odd, 
       'even' otherwise
    author
    date
    :type number: Int
    :rtype: String
    """
    # use modulo to check if the input is divisible by 2
    if number % 2 == 0:
        # if it is divisible by 2, then input is not odd
        return 'even'
    else:
        return 'odd'
    
is_odd(3)
```

# **Operadores Ternarios**

En python podemos utilizar el valor de ``True`` o ``False`` para evaluar condicionales

```python
estudiante= True

person = 'student' if estudiante else 'not student'
print(person)
```

El código anterior es equivalente a lo siguiente:

```python
estudiante= True

if estudiante:
    person = 'student'
else:
    person = 'not student'
        
print(person)
```
