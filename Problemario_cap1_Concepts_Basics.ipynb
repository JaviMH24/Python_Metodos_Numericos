3. Compute the slope between the points (3,4) and (5,9)
 - Recall that the slope between points (x1,y1) and (x2,y2) is $$\frac{y_2−y_1}{x_2−x_1}$$


```python
import numpy as np

x1, y1=3,4
x2, y2= 5,9

slope=(y2-y1)/(x2-x1)

print(slope)
```

4. Compute the distance between the points (3,4) and (5,9)
- Recall that the distance between points in two dimensions is $$\sqrt{(y_2−y_1)^2 + (x_2−x_1)^2}$$

```python
import numpy as np

x1, y1=3,4
x2, y2= 5,9


distance=np.sqrt((y2-y1)**2*(x2-x1)**2)

print(distance)
```
